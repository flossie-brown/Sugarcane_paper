```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
daily = np.load('old stuff/daily_sao_paulo.npy')
hourly = np.load('old stuff/hourly_sao_paulo.npy')
```

```python
fig, axs = plt.subplots(1,1, figsize = (9,9))

ax = axs
 
cf = ax.plot(daily, hourly, 'kx', markersize = 10)
ax.plot((0,140), (0,140), '--', color='grey', linewidth=3, alpha=0.4)

ax.plot((0,140), (0,140*1.39), '-', color='orange', linewidth=2, alpha=0.4)


plt.title('POD0 (1 year)', size = 25, pad = 20)
ax.set_xlabel('POD0 with monthly mean O$_3$ / mmol m$^{-2}$')
ax.set_ylabel('POD0 with hourly O$_3$ / mmol m$^{-2}$')

ax.set_xlim(0,140)
ax.set_ylim(0,140)

#plt.savefig('monthly_v_hourly.png')
```
